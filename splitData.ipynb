```python
import pandas as pd
from sklearn.model_selection import train_test_split
import os

# مسیر دیتاست
file_path = r"c:/Users/Hooshmand/Desktop/Dataset_Payanname/nosql/Final/synthetic_data_2000_binary_label-from400data.csv"

# خواندن داده
df = pd.read_csv(file_path)

# نام ستون لیبل (در صورت نیاز تغییر بده)
label_column = "label"

# تقسیم 50/50 با حفظ توزیع لیبل
train_df, test_df = train_test_split(
    df,
    test_size=0.2,
    stratify=df[label_column],
    random_state=42
)

# مسیر ذخیره (همان مسیر فایل اصلی)
save_dir = os.path.dirname(file_path)

train_path = os.path.join(save_dir, "synthetic_data_2000_binary_label-from400datatrain80.csv")
test_path = os.path.join(save_dir, "synthetic_data_2000_binary_label-from400datatest20.csv")

# ذخیره
train_df.to_csv(train_path, index=False)
test_df.to_csv(test_path, index=False)

print("Saved:")
print(train_path)
print(test_path)

# بررسی توزیع لیبل
print("\nTrain label distribution:")
print(train_df[label_column].value_counts(normalize=True))

print("\nTest label distribution:")
print(test_df[label_column].value_counts(normalize=True))
```
